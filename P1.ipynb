```python
def Normal_Interpolate(dados):
    
    # Calculando a média e o desvio padrão dos dados não nulos
    media = np.mean(dados[dados != 0])
    desvio_padrao = np.std(dados[dados != 0])
    
    interpolado = dados.copy()

    # Preenchendo os dados nulos com valores aleatórios de uma distribuição normal
    for i in range(len(dados)):
        if dados[i] == 0:
            interpolado[i] = np.random.normal(media, desvio_padrao)
            
    return interpolado
```
